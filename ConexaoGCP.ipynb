```python
'''
nomes, cpfs, pesos, alturas
'''

import pandas as pd

nomes = [
    'Aldo P', 'Alessandra', 'Amélia', 'Gabriela', 'Glauciamara',
    'Haroldo', 'Junior', 'Luciane', 'Martin', 'Neide'
]

cpfs = [
    '00011122233', '1112223344', '22233344455', '33344455566', '44455566677',
    '55566677788', '99988877766', '77766655544', '12312312322', '32132132111'
]

pesos = [
    77.6, 67.5, 87.4, 45.6, 76.7,
    88.9, 98.6, 67.6, 77.6, 66.7
]

alturas = [
    1.78, 1.67, 1.88, 1.77, 1.74,
    1.68, 1.78, 1.65, 1.68, 1.76
]

# Montar o dicionario de valores
dicionario_pessoas = {
    'nome': nomes,
    'cpf': cpfs,
    'peso': pesos,
    'altura': alturas
}

# criar um DataFrame pd.DataFrame(aquilo que você que vire um dataframe)
df_pessoas = pd.DataFrame(dicionario_pessoas)
display(df_pessoas)
```
