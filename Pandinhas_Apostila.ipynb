# Series e DataFrame

````python
# importar a biblioteca do pandas
import pandas as pd

# possui duas estruturas de dados
# Series - Estrutura unidimensional, similar a uma array com rótulos (index, indices)
# associativo

vendas = [2500, 3500, 3000, 4000, 2500, 4500]
meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho']

series_vendas = pd.Series(vendas) # Series criada sem um index, usando index automatico
display(series_vendas)
````

````python
# Atribuir um indice a minha Series
# index=lista - determina que o indice de cada valor da séries, será os elementos da lista.
series_vendas_meses = pd.Series(vendas, index=meses) # Series criada com um index associativo, sendo ele meses
display(series_vendas_meses)
````

````python
# Calcular o total de vendas do período
total_vendas = series_vendas_meses.sum() # Realiza a soma dos valores da Series
print(f"O total de vendas no Período foi R$ { total_vendas }")
````

## Novas linhas em medidas e algumas funções e calculos

````python
# Adicionar medidas extras ao DATAFRAME medidas

# Mediana
# median() - realiza o calculo da mediana
# Para adicionar novas linhas ao Dataframe pode usar a função loc[]
# no exemplo temos 3 colunas (idade, altura, peso), mediana de cada uma
# medidas.loc['median'] - qual o nome da linha
medidas.loc['mediana'] = [df['idade'].median(), df['altura'].median(), df['peso'].median()]
display(medidas)
````

````python
# Calcular a variancia (medida de dispersão)
# A variancia de cada coluna (idade, altura, peso).
# var()
medidas.loc['variancia'] = [df['idade'].var(), df['altura'].var(), df['peso'].var()]
display(medidas)
````

````python
# calcular a moda (valor que mais aparece)
# a moda de cada coluna, mode(), retorna o mais comum, mode()[0] vai ser o primeiro valor da moda retornada
# (idade, altura, peso)
medidas.loc['moda'] = [df['idade'].mode()[0], df['altura'].mode()[0], df['peso'].mode()[0]]
display(medidas)
````

````python
# Calculo do AIQ (Amplitude Interquartil) Q3 - Q1
# display(medidas.loc['75%'])
# display(medidas.loc['25%'])

# Variavel aiq = Q3 - Q1
medidas.loc['aiq'] = medidas.loc['75%'] - medidas.loc['25%']
display(medidas)
````

````python
# Cálculo dos limites inferiores e superiores, baseado no AIQ

# Limite inferior (LI): Q1 - (1.5 * AIQ)
medidas.loc['li'] = medidas.loc['25%'] - (1.5 * medidas.loc['aiq'])
# Limite superior (LS): Q3 + (1.5 * AIQ)
medidas.loc['ls'] = medidas.loc['75%'] + (1.5 * medidas.loc['aiq'])

display(medidas)
````

# Distribuição de Frequência

````python
import pandas as pd

dados = {
    'sexo': ['Feminino', 'Masculino', 'Feminino', 'Feminino', 'Masculino',
             'Masculino', 'Feminino', 'Masculino', 'Feminino', 'Masculino'],
    'idade': [23, 45, 35, 27, 52, 46, 29, 33, 41, 38],
    'renda': [3500, 7800, 5000, 4200, 8500, 8000, 4700, 5200, 6200, 7500],
    'escolaridade': ['Superior', 'Médio', 'Superior', 'Superior', 'Médio',
                     'Superior', 'Médio', 'Superior', 'Médio', 'Superior']
}

df = pd.DataFrame(dados)
display(df)
````

## Variáveis Qualitativas

````python
# Frequencia (Contagem)

# Sexo
# value_counts - realiza a contagem dos elementos dentro de uma coluna
freq_sexo = df['sexo'].value_counts()
display(freq_sexo)
````

````python
# Frequencia Relativa (%)

# Sexo
# value_counts(normalize=True) realiza a contagem relativa.
# retorna os valores como no exemplo, 0.5.
freq_sexo_rel = df['sexo'].value_counts(normalize=True) * 100
display(freq_sexo_rel)
````

````python
# Criar um DataFrame para exibir a tabelinha montada.
dist_freq_sexo = pd.DataFrame({
    'frequencia': freq_sexo,
    'frequencia_relativa_%': freq_sexo_rel
})

display(dist_freq_sexo)
````

````python
# Distribuição de Frequencia de Escolaridade.

# contagem
freq_escolaridade = df['escolaridade'].value_counts()
# relativa
freq_rel_escolaridade = df['escolaridade'].value_counts(normalize=True) * 100
# montar o DF
dist_freq_escolaridade = pd.DataFrame({
    'frequencia': freq_escolaridade,
    'frequencia_relativa_%': freq_rel_escolaridade
})
display(dist_freq_escolaridade)
````

````python
# Variáveis Quantitativas
# Faixas de Valores

import pandas as pd

dados = {
    'sexo': ['Feminino', 'Masculino', 'Feminino', 'Feminino', 'Masculino',
             'Masculino', 'Feminino', 'Masculino', 'Feminino', 'Masculino'],
    'idade': [23, 45, 35, 27, 52, 46, 29, 33, 41, 38],
    'renda': [3500, 7800, 5000, 4200, 8500, 8000, 4700, 5200, 6200, 7500],
    'escolaridade': ['Superior', 'Médio', 'Superior', 'Superior', 'Médio',
                     'Superior', 'Médio', 'Superior', 'Médio', 'Superior']
}

df = pd.DataFrame(dados)
display(df)
````

````python
# Adição de coluna Faixa de idade

# limites
bins_idade = [20, 30, 40, 50, 60]
# rótulos
labels_idade = ['20-29', '30-39', '40-49', '50-59']

# pd.cut é uma função do Pandas que categoriza os dados numéricos em faixas predefinidas.
# Argumentos:
# df['idade']: a coluna que será categorizada.
# bins=bins_idade: especifica as faixas de valores a serem usadas.
# labels=labels_idade: define os rótulos que correspondem a cada faixa.
# right=False: exclui o limite superior do intervalo, então uma idade de 30 estará na faixa '20 - 29'.
df['faixa_idade'] = pd.cut(df['idade'], bins=bins_idade, labels=labels_idade, right=False)
display(df)
````

````python
# Distribuição Quanti
# sort_index() realiza a ordenacao por indice
# contagem
freq_faixa_idade = df['faixa_idade'].value_counts().sort_index()
# relativa
freq_rel_faixa_idade = (df['faixa_idade'].value_counts(normalize=True) * 100).sort_index()

# frequencia acumulada
# cumsum() retorna a soma acumulada
# usa-se a variavel que contem a contagem como referencia
freq_acumulada_faixa_idade = freq_faixa_idade.cumsum()

# frequencia acumulada relativa
# use-se a variavel que contem relativa
freq_acumulada_rel_faixa_idade = freq_rel_faixa_idade.cumsum()

# Criar o DataFrame das frequencias
dist_faixa_idade = pd.DataFrame({
    'frequencia': freq_faixa_idade,
    'frequencia_%': freq_rel_faixa_idade,
    'frequencia_acumulada': freq_acumulada_faixa_idade,
    'frequencia_acumulada_%': freq_acumulada_rel_faixa_idade
})
display(dist_faixa_idade)
````

````python
# Frequencia de Renda
# bins_renda define as faixas de valores, que são [3000, 5000, 7000, 9000].
# labels_renda especifica os rótulos das faixas: '3000 - 4999', '5000 - 6999', '7000 - 8999'.
````

````python
bins_renda = [3000, 5000, 7000, 9000]
labels_renda = ['3000 - 4999', '5000 - 6999', '7000 - 8999']

df['faixa_renda'] = pd.cut(df['renda'], bins=bins_renda, labels=labels_renda, right=False)
# display(df)

# contagem
f_renda = df['faixa_renda'].value_counts().sort_index()
# relativa
fr_renda = (df['faixa_renda'].value_counts(normalize=True) * 100).sort_index()
# frequencia acumulada
fa_renda = f_renda.cumsum()
# frequencia acumulada relativa
far_renda = fr_renda.cumsum()

d_faixa_renda = pd.DataFrame({
    'frequencia': f_renda,
    'frequencia_%': fr_renda,
    'frequencia_acumulada': fa_renda,
    'frequencia_acumulada_%': far_renda
})

# display
display(d_faixa_renda)
````

# Agrupamento

````python
import pandas as pd
# Agrupamento (group by)

dados_vendas = {
    'categoria': ['Eletrônicos', 'Eletrônicos', 'Móveis', 'Móveis', 'Eletrônicos', 'Eletrodomésticos', 'Eletrodomésticos'],
    'subcategoria': ['Celulares', 'Televisores', 'Cadeiras', 'Mesas', 'Computadores', 'Geladeiras', 'Fogões'],
    'quantidade': [10, 5, 15, 7, 3, 8, 12],
    'valor_total': [5000, 12000, 3000, 7000, 15000, 16000, 10000]
}

df_vendas = pd.DataFrame(dados_vendas)
display(df_vendas)
````

````python
# dataframe_usado.groupby(por quem vc vai agrupar)[variavel quanti].operacao()
# operacao sum, max, min, mean, count

# Agrupar por Categoria e somar os valores de valor_total
agrupado = df_vendas.groupby('categoria')['valor_total'].sum()
agrupado
````

````python
# Agrupar por Categoria e calcular a media de valor_total
agrupado = df_vendas.groupby('categoria')['valor_total'].mean().round(2)
agrupado
````

````python
# Agrupar por subcategoria, e encontrar o valor maximo de quantidade
agrupado = df_vendas.groupby('subcategoria')['quantidade'].max()
agrupado
````

````python
# Agrupar por subcategoria, e encontrar o valor minimo de valor_total
agrupado = df_vendas.groupby('subcategoria')['valor_total'].min()
agrupado
````

````python
# Realizar agrupamentos por mais de uma dimensão (quali categorica)
# Agrupamento por categoria e subcategoria e somar o valor total
agrupado = df_vendas.groupby(['categoria', 'subcategoria'])['valor_total'].sum()
agrupado
````

````python
# Agrupamento, multiplas funcoes de agregacao (soma,media, min)
# agg(['', '', ''])
agrupado = df_vendas.groupby('categoria')['valor_total'].agg(['mean', 'max', 'min']).round(2)
agrupado
````

````python
# importar o mysql.connector e pandas
import mysql.connector as mc # apelido para usar os métodos do mysql.connector
import pandas as pd          # apelido para usar os métodos do pandas

# Definir variáveis de conexao
host = '34.44.188.226' # Endereço do servidor MySQL (IP Público da instancia) NAO CONFUNDA COM O IP DO COLAB!!!!
user = 'root'         # Nome de usuário do MySQL
password = ''         # Senha do MySQL (deve ser definida para a conexao)
database = 'consultas_db'  # Nome do banco de dados a ser utilizado
````

````python
# Descobre o IP do COLAB
!curl ipecho.net/plain
````

````python
# Estabelecer a conexao com o banco de dados
# parametroFunc=variavelConex
connection = mc.connect(
    host=host,
    user=user,
    password=password,
    database=database
)
# definindo o garçom
# o responsavel por executar as consultas do SQL, da conexao
cursor = connection.cursor()

# o garçom pede pra cozinha
# Executa a consulta passada como argumento da funcao
cursor.execute(
  '''
  SELECT *
  FROM personagens;
  '''
)

# o garçom recebe o pedido
# Recupera os dados da consulta
result = cursor.fetchall()

# o cliente saboreia o prato
# Converte o resultado para dataframe
# result - resultado da consulta, cursor.column_names - as colunas que retornaram na consulta
df = pd.DataFrame(result, columns=cursor.column_names)

# O garçom encerra o serviço
# Fechamento do cursor, após a consulta
cursor.close()

# A cozinha fecha o serviço
# Encerra a conexao do mysql
connection.close()

display(df.head())
````
